```python
mask = gdf['Oolloo'] != 0

gdf_O = gdf[mask]
```

```python
# FInd the distance from the AEM points to each point

gdf_O['nearest_ind'] = -999
gdf_O['nearest_distance'] = -999.

count = 0
for index, row in gdf_O.iterrows():
    dist = gdf_AEM.distance(row['geometry'])
    
    gdf_O.at[index,'nearest_ind'] = dist.idxmin()
    gdf_O.at[index, 'nearest_distance'] = dist.min()
    print(count)
    count+=1
```

```python
gdf_O['easting'] = gdf_O['geometry'].x
gdf_O['northing'] = gdf_O['geometry'].y
```

```python
# Remove points that are too far away
gdf_O_subset = gdf_O[gdf_O['nearest_distance'] < 500.]
```
