```python
import geopandas as gpd
import numpy as np
import os
from pyproj import Proj, transform

```

```python
infile = r"C:\Users\PCUser\Desktop\NSC_data\data\vector\HowardsEast\ProbabalisticInversionPts.shp"

gdf = gpd.read_file(infile)

# Create easting a northing column

inProj = Proj(init='epsg:4283')
outProj = Proj(init='epsg:28352')

gdf['X'] = np.nan
gdf['Y'] = np.nan

count = 0
for x in gdf['Long']:
    for y in gdf['Lat']:
        x2, y2 = transform(inProj,outProj,x,y)
        gdf.at[count, 'X'] = x2
        gdf.at[count, 'Y'] = y2
        
    count += 1
```

```python
infile = r"C:\Users\PCUser\Desktop\2017_HowardsEast_SkyTEM\01_EM\inversion_ready\HE_inversion_ready.dat"

aem_coords = np.loadtxt(infile, usecols = [5, 6])
```

```python
# Get the nearest neighbour

min_inds = -999 * np.ones(shape = len(gdf), dtype = np.int)

for i in range(len(gdf)):
    coords = gdf.iloc[i][['X', 'Y']].values
    min_inds[i] = np.argmin(np.abs(aem_coords - coords).sum(axis = 1))

#min_inds = np.sort(min_inds)
```

```python
# Now extract the line and write it into 

outfile = r"C:\Users\PCUser\Desktop\2017_HowardsEast_SkyTEM\01_EM\inversion_ready\HE_inversion_ready_Snowy_subset.dat"

with open(infile, 'r') as f:
    with open(outfile, 'w') as outf:
        for i, line in enumerate(f):
            if i in min_inds:
                outf.write(line)
                
        
```
