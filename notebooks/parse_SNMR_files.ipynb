```python
import pandas as pd
import numpy as np
import shutil
import os
import numpy
from shapely.geometry import Polygon
```

```python
def file2list(infile):
    L = []

    with open(infile, 'r') as f:
        for line in f:
            l = line.split()
            L.append(l)
    return L
```

```python
infile = r"E:\SSC\NMR\SNMR\SSC_SNMR_location_metadata.csv"

df = pd.read_csv(infile)

df['date'] = ''

df['bo_dec_angle_deg'] = ''
df['bo_inc_angle_deg'] = ''
df['geometry'] = ''
```

```python
dir = 'E:\\SSC\\NMR\\SNMR\\metadata_sheets'
```

```python
Polygon(geometry).wkt
```

```python
for index, row in df.iterrows():
    infile = os.path.join(dir, row['Field_ID'] + '_all_data.txt')
    L = file2list(infile)
    print(row['Field_ID'])
    geometry = np.zeros(shape = (4,2), dtype = np.float64)
    
    for l in L:
        if len(l) > 0:
            if l[0] == 'Date':
                df.at[index, 'date'] = l[2]
            if l[0] == 'Field':
                df.at[index, 'bo_dec_angle_deg'] = l[7]
                df.at[index, 'bo_inc_angle_deg'] = l[8]
            if len(l) > 1:
                try:
                    l.remove("Start")
                except ValueError:
                    pass
                if l[1] == 'SE':
                    geometry[0] = np.array([float(l[2]), float(l[3])])
                elif l[1] == 'SW':
                    geometry[1] = np.array([float(l[2]), float(l[3])])
                elif l[1] == 'NW':
                    geometry[2] = np.array([float(l[2]), float(l[3])])
                elif l[1] == 'NE':
                    geometry[3] = np.array([float(l[2]), float(l[3])])
                elif l[1] == 'Centre':
                    df.at[index, 'X'] = float(l[2])
                    df.at[index, 'Y'] = float(l[3])
    df.at[index, 'geometry'] = Polygon(geometry).wkt
```
